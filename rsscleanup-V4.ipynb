## HTML REMOVER FUNCTION ##

```python
def html_remover(raw_html):
    cleantext = BeautifulSoup(raw_html, "lxml").text
    cleantext = cleantext.replace('\n',' ')
    cleantext = cleantext.replace('\t', ' ')
    cleantext = cleantext.replace('\xa0',' ')
    return cleantext
```
